## Prepare the text for topic modeling

After removing blank complaints:

* Make the text lowercase
* Remove text in square brackets
* Remove punctuation
* Remove words containing numbers


After all the cleaning operations, performed the following:
* Lemmatize the texts
* Extract the POS tags of the lemmatized text and remove all the words which have tags other than NN[tag == "NN"].


```python
def clean_data(text: str) -> str:
    """
    Function to clean the text and remove unnecessary elements.

    Args:
        text (str): The input text to be cleaned.

    Returns:
        str: The cleaned text with unnecessary elements removed.
    """
    try:
        # Regular expressions for text cleaning
        cleaning_patterns = [
            (r'\s\{\$\S*', ''),      # Remove text within curly braces
            (r'\n', ''),             # Remove line breaks
            (r'\(\w*\)', ''),        # Remove text within braces
            (r'(\W\s)|(\W$)|(\W\d*)', ' '),   # Remove punctuation and special characters
            (r'x+((/xx)*/\d*\s*)|x*', ''),    # Remove date-like patterns
            (r'\d+\s', ''),          # Remove other numerical values
            (r' +', ' '),            # Remove unnecessary white spaces
        ]

        # Convert text to lowercase
        text = text.lower()

        # Apply each cleaning pattern
        for pattern, replacement in cleaning_patterns:
            text = re.sub(pattern, replacement, text)

        return text.strip()  # Remove leading and trailing spaces

    except Exception as e:
        # Log the error or raise a custom exception if necessary
        raise e
```

```python
# Apply data cleaning to the complaints_what_happened column

df['complaints'] = df['complaints_what_happened'].apply(clean_data)
df.head()
```

```python
def lemmatization(text: str) -> str:
    """
    Function to lemmatize the input text using the WordNetLemmatizer from NLTK.

    Args:
        text (str): The input text to be lemmatized.

    Returns:
        str: The lemmatized text.
    """
    try:
        # Tokenize the text into words
        tokens = word_tokenize(text)

        # Initialize WordNetLemmatizer
        wordnet_lemmatizer = WordNetLemmatizer()

        # Lemmatize each token in the text
        lemmatized = [wordnet_lemmatizer.lemmatize(token) for token in tokens]

        # Join the lemmatized words into a string
        lemmatized_str = " ".join(lemmatized)

        return lemmatized_str

    except Exception as e:
        # Log the error or raise a custom exception if necessary
        raise e
```

```python
#Write your function to extract the POS tags 

def get_pos_tags(text: str) -> str:
    """
    Function to extract the lemmatized text with only nouns (NN) from the input text.

    Args:
        text (str): The input text to be processed.

    Returns:
        str: The lemmatized text containing only nouns (NN).
    """
    try:
        # Process the text with spaCy NLP model
        doc = nlp(text)

        # Extract lemmatized nouns (NN)
        nn_words = [tok.lemma_ for tok in doc if tok.tag_ == 'NN']

        # Join the lemmatized words into a string
        nn_words_str = " ".join(nn_words)

        return nn_words_str

    except Exception as e:
        # Log the error or raise a custom exception if necessary
        print("Error: Failed to process the text with spaCy NLP model.")
        raise e
```

```python
df_clean["complaint_POS_removed"] =  df_clean.swifter.apply(lambda x: get_pos_tags(x['lemmatized']), axis=1)
```

#### Find the top unigrams,bigrams and trigrams by frequency among all the complaints after processing the text.

```python
def top_grams(grams, df_clean):
    """
    Function to find the top n-grams (words or phrases) frequency-wise in the given DataFrame.

    Parameters:
        grams (int): The n in n-grams (e.g., 1 for unigrams, 2 for bigrams, etc.).
        df_clean (pd.DataFrame): The DataFrame containing the cleaned complaints.

    Returns:
        pd.DataFrame: A DataFrame with the top n-grams and their frequencies.
    """
    # Create a CountVectorizer to generate n-grams
    c_vec = CountVectorizer(stop_words=stopwords.words('english'), ngram_range=(grams, grams))

    # Fit and transform the cleaned complaints to get the n-grams frequencies
    grams_data = c_vec.fit_transform(df_clean['complaint_POS_removed'])

    # Calculate the total count of each n-gram
    count_values = grams_data.toarray().sum(axis=0)

    # Get the vocabulary and corresponding n-grams
    vocab = c_vec.vocabulary_
    ngram_list = sorted([(count_values[i], k) for k, i in vocab.items()], reverse=True)

    # Create a DataFrame to store the n-grams and their frequencies
    df_ngram = pd.DataFrame(ngram_list, columns=['frequency', 'ngram'])

    return df_ngram
```

## Feature Extraction
Convert the raw texts to a matrix of TF-IDF features

**max_df** is used for removing terms that appear too frequently, also known as "corpus-specific stop words"
max_df = 0.95 means "ignore terms that appear in more than 95% of the complaints"

**min_df** is used for removing terms that appear too infrequently
min_df = 2 means "ignore terms that appear in less than 2 complaints"

```python
#Write your code here to initialise the TfidfVectorizer 
tfidf = TfidfVectorizer(max_df=0.95, min_df=2, stop_words='english')
```

#### Create a document term matrix using fit_transform

The contents of a document term matrix are tuples of (complaint_id,token_id) tf-idf score:
The tuples that are not there have a tf-idf score of 0

```python
#Write your code here to create the Document Term Matrix by transforming the complaints column present in df_clean.
dtm = tfidf.fit_transform(df_clean['complaints'])
```

## Topic Modelling using NMF

Non-Negative Matrix Factorization (NMF) is an unsupervised technique so there are no labeling of topics that the model will be trained on. The way it works is that, NMF decomposes (or factorizes) high-dimensional vectors into a lower-dimensional representation. These lower-dimensional vectors are non-negative which also means their coefficients are non-negative.

In this task we will perform the following:

* Find the best number of clusters 
* Apply the best number to create word clusters
* Inspect & validate the correction of each cluster wrt the complaints 
* Correct the labels if needed 
* Map the clusters to topics/cluster names

```python
from sklearn.decomposition import NMF
```

## Manual Topic Modeling
We need to do take the trial & error approach to find the best num of topics for your NMF model.

The only parameter that is required is the number of components i.e. the number of topics we want. This is the most crucial step in the whole topic modeling process and will greatly affect how good our final topics are.

```python
#Load your nmf_model with the n_components i.e 5
num_topics = 5

#keep the random_state =40
nmf_model = NMF(random_state=40, n_components=num_topics)
```

```python
nmf_model.fit(dtm)
```

```python

len(tfidf.get_feature_names_out())
```

```python
#Print the Top15 words for each of the topics
words = np.array(tfidf.get_feature_names_out())
topic_words_df = pd.DataFrame(np.zeros((num_topics, 15)), index=[f'Topic {i + 1}' for i in range(num_topics)],
                           columns=[f'Word {i + 1}' for i in range(15)]).astype(str)

for i in range(num_topics):
    ix = nmf_model.components_[i].argsort()[::-1][:15]
    topic_words_df.iloc[i] = words[ix]

topic_words_df
```

```python
#Create the best topic for each complaint in terms of integer value 0,1,2,3 & 4
topic_results = nmf_model.transform(dtm)
topic_results.argmax(axis=1)
```

#### Applying the supervised models on the training data created. In this process, we are going to do the following:
* Create the vector counts using Count Vectoriser
* Transform the word vecotr to tf-idf
* Create the train & test data using the train_test_split on the tf-idf & topics


```python
# Write your code to get the Vector count
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(training_data.complaints)

# Write your code here to transform the word vector to tf-idf
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
```

You have to try atleast 3 models on the train & test data from these options:
* Logistic regression
* Decision Tree
* Random Forest
* Naive Bayes (optional)

**Using the required evaluation metrics judge the tried models and select the ones performing the best**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import classification_report
```

```python
# Split data into train and test
X_train, X_test, y_train, y_test = train_test_split(X_train_tfidf, training_data.Topic, test_size=0.2, random_state=42)
```

**1. Logistic Regression**

```python
# Logistic Regression
lr = LogisticRegression().fit(X_train, y_train)
predicted = lr.predict(X_test)

print(classification_report(y_true=y_test, y_pred=predicted))
```

**2. Decision Tree**

```python
# Decision tree classifier
dt = DecisionTreeClassifier().fit(X_train, y_train)
predicted = dt.predict(X_test)

print(classification_report(y_pred=predicted, y_true=y_test))
```

**3. Random Forest Classifier**

```python
rfc = RandomForestClassifier(max_depth=10)
rfc.fit(X_train, y_train)
predicted = rfc.predict(X_test)

print(classification_report(y_pred=predicted, y_true=y_test))
```

**4. Gaussian Naive Bayes**

```python
nb = GaussianNB().fit(X_train.toarray(), y_train)
predicted = nb.predict(X_test.toarray())

print(classification_report(y_pred=predicted, y_true=y_test))
```

```python
def predict_lr(text, count_vect, tfidf_transformer, lr):
    """
    Function to predict the topic of a given text using a trained Logistic Regression model.

    Parameters:
        text (str): The input text to be classified.
        count_vect (CountVectorizer): The CountVectorizer used for transforming text to word counts.
        tfidf_transformer (TfidfTransformer): The TfidfTransformer used for transforming word counts to TF-IDF scores.
        lr (LogisticRegression): The trained Logistic Regression model.

    Returns:
        str: The predicted topic label for the given text.
    """
    Topic_names = {
        0: 'Theft/Dispute Reporting',
        1: 'Credit card or prepaid card',
        2: 'Mortgage/Loan',
        3: 'Account Services',
        4: 'Others'
    }

    try:
        # Transform the input text using CountVectorizer and TfidfTransformer
        X_new_counts = count_vect.transform([text])
        X_new_tfidf = tfidf_transformer.transform(X_new_counts)

        # Predict the topic using the Logistic Regression model
        predicted = lr.predict(X_new_tfidf)[0]

        # Return the predicted topic label
        return predicted

    except Exception as e:
        print("Error: Failed to predict the topic.",e)
        return None

# Example usage
text_to_predict = "I am having issues with my credit card statement."
predicted_topic = predict_lr(text_to_predict, count_vect, tfidf_transformer, lr)
print(predicted_topic)
```
